# The Joint Plot
It displays the relation between two variables and also their data distribution

```python
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

# Plain Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df)
```

# Joint Plot with Regression Line
The shaded area around the regression line indicates the area where true value may lie. The wider the shade, the more uncertain or inaccurate regression line.

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```

# Joint Plot with Hexagonal points
The darker hexagons represent more density of points

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='hex')
```
