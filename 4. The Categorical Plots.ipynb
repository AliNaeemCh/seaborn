# The Categorical Plots
The categorical plots display different stats of the numerical data like mean, median, sum, standard deviation etc for different categories

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

### The Bar Plot
By default, the estimator is mean which means on the y-axis the mean will be displayed. You can change it to sum or standard deviation etc.

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.sum, errorbar=None)
plt.title('Accumulated Money Spent By Males and Females')
plt.show()
```

### The Count Plot

```python
sns.countplot(x='sex', data=tips_df)
plt.title('Male vs Female Customers')
plt.show()
```

### The Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.title('Range of Bills vs Week Days')
plt.show()
# The spots are the outliers which aren't included in the calculation
```

### The Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', dodge=True)
plt.title('Bills vs Week Days')
plt.show()
# Dodge=True, separates the points of male and female
```
