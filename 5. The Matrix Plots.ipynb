# The Matrix Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
numerical_columns = tips_df.select_dtypes(include=['number'])
```

### The Heat Map

```python
corr = numerical_columns.corr()
corr
# Note that a correlation of 1 means linearly related.
# A correlation of -1 means inversely related
# And a correlation of 0 means no linear correlation
```

```python
sns.heatmap(corr, annot=True, cmap='Blues', linecolor='white', linewidth=2)
```

### The Pair Grid Plot
In pair grid plot, you can specify the variables you want to compare

```python
grid = sns.PairGrid(tips_df, hue='sex', x_vars=['total_bill', 'tip'], y_vars=['total_bill', 'tip'])
grid.map(plt.scatter)
grid.add_legend()
```

### The Facet Grid Plot
In this matrix plot, you can give a categorical column and specify two variables to compare and it will create subplots for all the categories

```python
fg = sns.FacetGrid(tips_df, col='time', hue='smoker', palette='Set1')
fg.map(plt.scatter, 'total_bill', 'tip')
fg.add_legend()
```
