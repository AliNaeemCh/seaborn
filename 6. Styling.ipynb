# Styling

```python
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```

# Changing Grid Style

```python
sns.set_style('darkgrid')
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```

# Adding Ticks

```python
sns.set_style('ticks')
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```

# Changing Size

```python
# By default, it is paper
sns.set_context('talk', font_scale=1.5)
# Font scale, scales all the writings
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```
